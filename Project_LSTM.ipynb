Fitting an LSTM model:

```python
# drop visit_id column as we have patient_id and visit_month columns
new_train_df = new_train_df.drop('visit_id', axis = 1)
```

```python
import pandas as pd

# Assuming df is your DataFrame containing the dataset and 'patient_id' is the column with patient IDs
# And 'k' is the value by which you want to filter the patient IDs

# Get unique patient IDs
unique_patient_ids = new_train_df['patient_id'].unique()

# Initialize an empty list to store filtered DataFrames
filtered_dfs = []

k = 3

import pandas as pd

# Assuming new_train_df is your DataFrame containing the dataset,
# unique_patient_ids is the list of unique patient IDs,
# and k is the value by which you want to filter the rows.

# Initialize an empty list to store filtered DataFrames
filtered_dfs = []

# Loop over unique patient IDs
for patient_id in unique_patient_ids:
    # Subset the DataFrame for the current patient ID
    subset_df = new_train_df[new_train_df['patient_id'] == patient_id]
    
    # Calculate the remainder when dividing the count of rows by 'k'
    remainder = len(subset_df) % k
    
    # Determine the number of excess rows
    excess_rows = len(subset_df) - (len(subset_df) // k) * k
    
    # Determine the number of rows to keep for this patient ID
    rows_to_keep = len(subset_df) - excess_rows
    
    # Keep rows with indices up to 'rows_to_keep'
    filtered_subset_df = subset_df.iloc[:rows_to_keep]
    
    # Append the filtered subset DataFrame to the list
    filtered_dfs.append(filtered_subset_df)

# Concatenate all filtered DataFrames into a single DataFrame
filtered_df = pd.concat(filtered_dfs)

# Now, filtered_df contains the original DataFrame with rows removed based on the specified remainder condition for each patient ID
```

```python
filtered_df.head(10)
```

```python
filtered_df.shape
```

```python
filtered_df['patient_id'].nunique()
```

```python
# creating sequences for LSTM
import pandas as pd
import numpy as np

# Assuming 'data' is your dataset and 'patient_id' is a column representing patient IDs

new_train_df = filtered_df

sequences = []  # List to store sequences
sequence_length = 3  # Define the length of each sequence

for patient_id in new_train_df['patient_id'].unique():
    # Subset the DataFrame for the current patient ID
    patient_data = new_train_df[new_train_df['patient_id'] == patient_id]
    
    # Determine the number of sequences to create
    num_sequences = max(len(patient_data) // sequence_length, 0)
    
    # Create sequences from patient data
    for i in range(num_sequences):
        start_index = i * sequence_length
        end_index = (i + 1) * sequence_length
        sequence = patient_data.iloc[start_index:end_index]
        sequences.append(sequence)
```

```python
import pandas as pd

# Form the target variable with patient id and updrs_1 for all visits
target = []
for i in range(len(sequences)):
    patient_id = sequences[i]['patient_id'].iloc[0]  # Get patient_id for the sequence
    updrs_1_values = sequences[i]['updrs_1'].tolist()  # Get updrs_1 values for all visits
    target.append({'patient_id': patient_id, 'updrs_1_visit_1': updrs_1_values[0], 'updrs_1_visit_2': updrs_1_values[1], 'updrs_1_visit_3': updrs_1_values[2]})

# Display the first target
print(target[0])

# Remove the scores from the sequences
for i in range(len(sequences)):
    sequences[i] = sequences[i].drop(['updrs_1'], axis=1)
```

```python
import pandas as pd
import numpy as np

# Convert the list of dictionaries into a pandas DataFrame
target_df = pd.DataFrame(target)

# Display the first few rows of the DataFrame
print(target_df.head())

# Convert the DataFrame into a NumPy array
target_array = target_df.to_numpy()

# Display the first element of the NumPy array
print(target_array[0])
```

```python
# train test split
import pandas as pd
from sklearn.model_selection import train_test_split

# Assuming 'sequences' contains your sequence data and 'target_values' contains the target variable
# Assuming 'patient_ids' is a column in your dataset representing patient IDs

# Get unique patient IDs
# Assuming sequences is a list of sequences, and each sequence contains a 'patient_id' attribute
unique_patients = set(sequence['patient_id'].iloc[0] for sequence in sequences)

import random

# Convert the set to a list, shuffle it, and convert back to a set
patients_list = list(unique_patients)
random.shuffle(patients_list)

# Select a fraction of patients for the test set
test_fraction = 0.2  # Set the fraction of patients you want to use for testing
num_test_patients = int(len(patients_list) * test_fraction)
test_patients = set(patients_list[:num_test_patients])

# Create the training set by subtracting the test set from the shuffled patients
train_patients = unique_patients - test_patients

print("Training Patients:", train_patients)
print("Test Patients:", test_patients)
```

```python
len(train_patients)
```

```python
len(test_patients)
```

```python
# check for common patients in train and test
common_patients = list(set(train_patients).intersection(set(test_patients)))
print(common_patients)
```

```python
# Initialize train and test lists for sequences and targets
train_sequences, test_sequences = [], []
train_target, test_target = [], []

# Iterate over sequences and target list
for sequence, target_entry in zip(sequences, target_array):
    patient_id = target_entry[0]
    updrs_1_visit_1 = target_entry[1]
    updrs_1_visit_2 = target_entry[2]
    updrs_1_visit_3 = target_entry[3]

    # Check if patient belongs to train or test set
    if patient_id in train_patients:
        train_sequences.append(sequence)
        train_target.append(np.array([updrs_1_visit_1, updrs_1_visit_2, updrs_1_visit_3]))
    elif patient_id in test_patients:
        test_sequences.append(sequence)
        test_target.append(np.array([updrs_1_visit_1, updrs_1_visit_2, updrs_1_visit_3]))
```

```python
# LSTM code
import tensorflow as tf
import keras.losses
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import LSTM, Dense
from sklearn.metrics import mean_squared_error
from sklearn.metrics import r2_score

# reshape the train and test sequences
train_sequences = np.array(train_sequences)
test_sequences = np.array(test_sequences)

# reshape the train and test target
train_target = np.array(train_target)
test_target = np.array(test_target)

# Define the model
model = Sequential()
model.add(LSTM(128, input_shape=(train_sequences.shape[1], train_sequences.shape[2]), return_sequences=True))
#model.add(LSTM(64, return_sequences=True))  # Last LSTM layer may not return sequences
model.add(LSTM(64))
model.add(Dense(3))  # Output layer for regression

# Compile the model
model.compile(optimizer='adam', loss=keras.losses.mean_absolute_error)  # Choose appropriate loss for regression

# Train the model
model.fit(train_sequences, train_target, epochs=100, batch_size=20, validation_data=(test_sequences, test_target))

# Make predictions on the test set
predictions = model.predict(test_sequences)
# training error
train_predictions = model.predict(train_sequences)
mae = model.evaluate(train_sequences, train_target)
print("Mean Absolute Error (MAE)of training :", mae)

# Calculate Mean Squared Error (MSE)
mae = model.evaluate(test_sequences, test_target)
print("Mean Absolute Error (MAE) of testing:", mae)
```
